Checking most mentioned genres

```python
categorical_column = 'Main Genre'

if categorical_column in book_df_1.columns:
    genre_counts = book_df_1[categorical_column].value_counts()

    plt.style.use('_mpl-gallery')

    fig, ax = plt.subplots(figsize=(16, 8))

    ax.bar(genre_counts.index, genre_counts.values, edgecolor="white")

    ax.set_ylabel("Frequency")
    ax.set_xlabel(categorical_column)
    ax.set_title(f"Distribution of {categorical_column}")

    plt.xticks(rotation=90, ha='right')

    plt.tight_layout() 
    plt.show()
else:
    print(f"Error: Column '{categorical_column}' not found in the DataFrame.")
```
